```python
c = []
for x in Genotype.columns[10:]:
    if 'Williams82_B01' == x:
        x = 'Williams82'
    c.append(namemap_ix.loc[x]['2018년 포장번호'])
```

```python
Genotype_sub = Genotype[Genotype.columns[10:]]
```

```python
Genotype_sub.columns = c
```

```python
Genotype_sub.head()
```

```python
def seq_convert(x):
    dicN = {'A':'1','T':'2','G':'3','C':'4','N':'0'}
    gt = list(x)
    gt.sort()
    result = ''.join([dicN[i] for i in x])
    return int(result,5)
seq_convert_vec = np.vectorize(seq_convert)
genotype_seq_vec = seq_convert_vec(np.stack(Genotype_sub.values))
df_genotype_vec = pd.DataFrame(genotype_seq_vec, columns=Genotype_sub.columns)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=50)
```

```python
data = pca.fit_transform(df_genotype_vec.T)
```

```python
genotype_pca = pd.DataFrame(data, index=Genotype_sub.columns)
```

```python
import scipy.cluster.hierarchy as sch
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
```

```python
X = genotype_pca.values
```

```python
dendrogram = sch.dendrogram(sch.linkage(X, method  = "ward"))
plt.title('genotype_pca')
plt.show()
```

```python
ix = [int(x) for x in dendrogram['ivl']]
```

```python
sample_name = sample_name[sample_name.columns[1:]]
```

```python
sample_name
```

```python
label_list = [sample_name[sample_name['SampleNo'] == x]['ResourceName'].values.tolist() for x in genotype_pca.T.columns[ix].tolist()]
```

```python
X = genotype_pca.values
labelList = sum(label_list,[])
linked = linkage(X, 'ward')
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', labels=labelList, distance_sort='descending', show_leaf_counts=True)
plt.savefig('CRS_hierarchy.png', dpi=800, bbox_inches='tight')
plt.show()
```
